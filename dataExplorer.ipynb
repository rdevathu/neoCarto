```python
import pandas as pd
```

```python
date_columns = [
    'Date of Birth', 'Procedure date', 'AF recurrence date', 
    'AT recurrence date', 'Date of repeat ablation?'
]

data = pd.read_csv('data.csv', parse_dates=date_columns)

pd.set_option('display.max_columns', None)

data.head()
```

```python
print(data.shape)
```

```python
columns_to_keep = [
    'Record ID', 'MRN', 'Date of Birth', 'Name', 'Procedure date', 'Procedure Type', 
    'PVI (+/- Carina and CTI) only', 'Type of AF (0=paroxysmal, 1=persistent)', 'First AF/AT ablation?', 'ECG every 3ish months?', 
    'AF recurrence after blanking period (60 days)?', 'AF recurrence date', 'Days till AF recurrence', 
    'AT recurrence after blanking period (60 days)?', 'AT recurrence date', 'Days till AT recurrence', 
    'Date of repeat ablation?'
]

existing_columns = [col for col in columns_to_keep if col in data.columns]
missing_columns = [col for col in columns_to_keep if col not in data.columns]

print(f"Existing columns: {len(existing_columns)}")
print(f"Missing columns: {missing_columns}")

filtered_data = data[existing_columns].copy()
print(f"\nFiltered data shape: {filtered_data.shape}")
```

```python
# Combined AF/AT recurrence: 1 if either af_recurrence or at_recurrence is 1
filtered_data['AF/AT Recurrence'] = (
    (filtered_data['AF recurrence after blanking period (60 days)?'] == 1) | (filtered_data['AT recurrence after blanking period (60 days)?'] == 1)
).astype(float)
both_missing = filtered_data['AF recurrence after blanking period (60 days)?'].isna() & filtered_data['AT recurrence after blanking period (60 days)?'].isna()
filtered_data.loc[both_missing, 'AF/AT Recurrence'] = float('nan')

# Combined recurrence date: earlier of the two recurrence dates
filtered_data['AF/AT Recurrence Date'] = filtered_data[['AF recurrence date', 'AT recurrence date']].min(axis=1)

# Combined days to recurrence: shorter time period
filtered_data['Days till AF/AT Recurrence'] = filtered_data[['Days till AF recurrence', 'Days till AT recurrence']].min(axis=1)

```

```python
count_first_ablation = filtered_data['First AF/AT ablation?'].notnull().sum()
print(f"Count of first_ablation (non-null): {count_first_ablation}")

# First time AF patients
count_first_time_af = filtered_data[filtered_data['Type of AF (0=paroxysmal, 1=persistent)'] == 0]['First AF/AT ablation?'].notnull().sum()
print(f"Count of first_ablation with type_of_af = 0 (first time AF): {count_first_time_af}")

# First time AF patients with PVI only
count_first_time_af_pvi_only = filtered_data[
    (filtered_data['Type of AF (0=paroxysmal, 1=persistent)'] == 0) & (filtered_data['PVI (+/- Carina and CTI) only'] == 1)
]['First AF/AT ablation?'].notnull().sum()
print(f"Count of first_ablation with type_of_af = 0 and PVI only = 1: {count_first_time_af_pvi_only}")
```

```python
# Analysis subset: first ablation patients with first-time AF and PVI only
subset = filtered_data[
    filtered_data['First AF/AT ablation?'].notnull() &
    (filtered_data['Type of AF (0=paroxysmal, 1=persistent)'] == 0) &
    (filtered_data['PVI (+/- Carina and CTI) only'] == 1)
]

total_patients = len(subset)

af_recurrence_count = subset[subset['AF recurrence after blanking period (60 days)?'] == 1].shape[0]
af_recurrence_pct = (af_recurrence_count / total_patients) * 100 if total_patients > 0 else 0

af_within_1yr = subset[(subset['AF recurrence after blanking period (60 days)?'] == 1) & (subset['Days till AF recurrence'] <= 365)]
af_within_1yr_count = af_within_1yr.shape[0]
af_within_1yr_pct = (af_within_1yr_count / total_patients) * 100 if total_patients > 0 else 0

print(f"Total patients (first ablation, type_of_af=0, PVI only=1): {total_patients}")
print(f"Count with AF recurrence = 1: {af_recurrence_count} ({af_recurrence_pct:.2f}%)")
print(f"  ...within 1 year: {af_within_1yr_count} ({af_within_1yr_pct:.2f}%)")

at_recurrence_count = subset[subset['AT recurrence after blanking period (60 days)?'] == 1].shape[0]
at_recurrence_pct = (at_recurrence_count / total_patients) * 100 if total_patients > 0 else 0

at_within_1yr = subset[(subset['AT recurrence after blanking period (60 days)?'] == 1) & (subset['Days till AT recurrence'] <= 365)]
at_within_1yr_count = at_within_1yr.shape[0]
at_within_1yr_pct = (at_within_1yr_count / total_patients) * 100 if total_patients > 0 else 0

print(f"Count with AT recurrence = 1: {at_recurrence_count} ({at_recurrence_pct:.2f}%)")
print(f"  ...within 1 year: {at_within_1yr_count} ({at_within_1yr_pct:.2f}%)")

afat_recurrence_count = subset[subset['AF/AT Recurrence'] == 1].shape[0]
afat_recurrence_pct = (afat_recurrence_count / total_patients) * 100 if total_patients > 0 else 0

afat_within_1yr = subset[(subset['AF/AT Recurrence'] == 1) & (subset['Days till AF/AT Recurrence'] <= 365)]
afat_within_1yr_count = afat_within_1yr.shape[0]
afat_within_1yr_pct = (afat_within_1yr_count / total_patients) * 100 if total_patients > 0 else 0

print(f"Count with AF/AT Recurrence = 1: {afat_recurrence_count} ({afat_recurrence_pct:.2f}%)")
print(f"  ...within 1 year: {afat_within_1yr_count} ({afat_within_1yr_pct:.2f}%)")

# export subset to csv
subset.to_csv('subset.csv', index=False)
```
